```python
from mlxtend.preprocessing import TransactionEncoder
encod = TransactionEncoder()
transaction = encod.fit(groc).transform(groc)
```

```python
transaction = transaction.astype('int')
```

```python
df = pd.DataFrame(transaction, columns=encod.columns_)
```

```python
df.head()
```

```python
df.info()
```

```python
from mlxtend.frequent_patterns import apriori

def freq(min_support):
  frequent = apriori(df, min_support=min_support, use_colnames=True)
  frequent['length'] = frequent['itemsets'].apply(lambda x: len(x))
  return frequent

min_support=float(input("Enter the minimum support: "))
fr=freq(min_support)
print(fr)
     
```

```python
fr.head()
```

```python
#Itemset having the itemset more than or equal to 2 and minimum support 5%
fr[(fr['length'] >= 2) & (fr['support'] >= 0.05)]
```

```python
from mlxtend.frequent_patterns import association_rules
con = association_rules(fr, metric='support', min_threshold=0.02)
```
